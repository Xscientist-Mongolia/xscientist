```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

Та өмнөх хэсгээс бит болон күбит гэж юу болох, болон хэрхэн бидний өдөр тутам хэрэглэдэг компьютер ажилладаг талаар мэдэж авсан. Орчин үеийн програманд (software) агуулагддаг зураг, дүрс, хувьсагч, ба өгөгдөлийн бүтэц нь ердөө л битүүд юм. Квант компьютер дээр ажиллах бид бүхэн дээрх зүйлсийг классик хувьсагч гэж нэрлэх болно. Түүнчлэн таны одоо хэрэглэж буй буюу энэ бичвэрийг уншиж буй компьютерийг классик компьютер гэж нэрлэе. Квант компьютерийн хувьд бидний үндсэн хувьсагч бол күбит юм: Битийн квант хуилбар. Күбит нь квант зүйл бөгөөд квант физикийн хуульд захирагддаг. Энэ хэсэгт бид бүхэн квант физикийн нэгэн чухал шинж болох $\textbf{Тодорхойгүйн зарчим}$-ын талаар авч үзэх болно.

## Хаязэнбэргийн тодорхойгүйн зарчим (Heisenberg's uncertainty principle) 

Тодорхойгүйн зарчимийн хамгийн ерөнхий тодорхойлолт бол бөөмийн байрлал болон моментийн хамаарлын тухай юм: бөөмийн байрлалын мэдээлэл их байх тусам, түүний моментийн мэдээлэл илүү тодорхойгүй болдог.   

```python
from IPython.display import Image
Image(filename='heisenberg.png')
```

Энэ бол квант обектийн хамгийн ерөнхий шинж чанар юм. Энд хувьсагч нь заавал байрлал мөн момент байх шаардлагагүй төдийгүй өөр олон тарлийн хувьсагчууд буюу параметрүүд байдаг. Өөр үгээр хэлбэл нэг хувьсагчийн мэдээлэл нь ихсэх тусам бусад хувьсагчийн мэдээлэл нь тодорхогүй буюу санамсаргүй болж эхэлдэг. Тодорхойгүйн зарчим хэрхэн күбитэд нөлөөлдөг вэ? гэдгийг сонирхож доорх хэмжилтийг хийе.

```python
measure_z=QuantumCircuit(1,1) #z-measurement нь күбитийг хэвээр үлдээдэг.
measure_z.measure(0,0)
measure_z.draw(output='mpl')
#энд дотроо цагаантай хар хайрцаг нь ямар нэгэн хэмжилт хийж буйг илэрхийлж байгаа,
#Харин доод талын 2 шугам (C) нь хэмжилтийн утгыг дэлгэцэнд гаргаж буйг илэрхийлж байна. 
```

Одоо 1-р хэсэгт дурдсан Х gate-ээр үйлчилэе.

```python
measure_x=QuantumCircuit(1,1)# x-measurement буюу квант NOT гэйт. 
measure_x.h(0)
measure_x.measure(0,0)
measure_x.draw(output='mpl')
```

Харин одоо хоосон хэлхээ сонирхоё.

```python
qc_0=QuantumCircuit(1)
qc_0.draw(output='mpl')
```

Күбитийг z хэмжилтйин дараа 0-г гаралтанд өгөхөөр анхны төлөвт нь бэлдсэн байдаг. Доорх үр дүнг харцгаая.

```python
qc=qc_0+measure_z
print('z хэмжилтийн гаралтын үр дүн:')
counts=execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

Хэрэв бид x хэмжилт гүйцэтгэвэл гаралтын үр дүн (өмнөх үр дүнгээс эсэргээрээ) санамсаргүй байх болно. 

```python
qc=qc_0+measure_x
print('x хэмжилтийн гаралтын үр дүн:')
counts=execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

Одоо Хадамард (Hadamard) гэйтийг үйлчилбэл юу болохыг сонирхоё.

```python
qc_plus=QuantumCircuit(1)
qc_plus.h(0)
qc_plus.draw(output='mpl')
```

Хадамард гэйт үйлчилсэний дараах z хэмжилтийн үр дүнг сонирхоё.

```python
qc=qc_plus+measure_z
qc.draw()

print('z хэмжилтийн гаралтын үр дүн:эндээс харвал үр дүн санамсаргүй (тодорхойгүй) байна гэдгийг харж болно.')
counts=execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

Харин тэгвэл одоо x хэмжилтийн үр дүн нь ямар байхыг сонирхоё.

```python
qc=qc_plus+measure_x

print('x хэмжилтийн гаралтын үр дүн: маш тодорхой байна гэдгийг харж болно.')
counts=execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

Бидний квант бит буюу күбит нь 1 хэмжээтэй мөн чиглэлтэй вектор буюу доорх байдлаар дүсэлбэл бид күбитийг тодорхой өнцгөөр эргүүлж болно гэдэг нь харагдаж байна. Күбитийг битээс ялгахын тулд $|0>$ $|1>$ гэж тэмдэглэдэг.

```python
from IPython.display import Image
Image(filename='bloch.jpg')
```

y тэнхэлгийг тойруулан тодорхой өнцгөөр эргүүлэе.

```python
qc_y=QuantumCircuit(1)
qc_y.ry(-3.14159/4,0)
qc_y.draw(output='mpl')
```

Одоо z болон x хэмжилтйин үр дүнг сонирхоё. 

```python
qc=qc_y+measure_z

print('Results for z measurement:')
counts = execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

```python
qc=qc_y+measure_x

print('Results for x measurement:')
counts = execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

Энэ 2 хэмжилтийн үр дүнгээс харвал, хэмжилтийн утга тодорхойгүй байна. Эндээс дүгнэвэл хэмжилтйин үр дүн тодорхой байх хязгаартай буюу хэмжилтийн үр дүн хэзээ ч 100 хувь тодорхой байж чадахгүй. Дэлгэрүүлбэл дээр харсанчлан x хэмжилтийн үр дүн 0 эсвэл 1 ойролцоо тэнцүү магадлалтай үр дүн өгөх, харин z хэмжилт 0 гаралтанд өгөх ёстой байсан.
