# Мэдээллийг бит (0/1)-ээр илэрхийлэх

Бит 0 болон 1-ээр дурын мэдээлийг илэрхийлж болдог. Дурын харилцан ярианы хэлэнд тоог 0-оос 9-ийн хоорондох (0,1,2,3,4,5,6,7,8 ба 9) цифрүүдээр илэрхийлдэг. Цифр бүр нь тухайн тоо хэдэн ширхэг 10-ийн зэрэгтйиг агуулж байгааг илэрхийлдэг. Жишээ болгож 9213 тоог сохирхож үзэе. Энэ тоог $ 9000+200+10+3 $ хэлбэртэй бичиж болох буюу хэрэв 10-тын зэргээр илэрхийлбэл  $ 9\times 10^3+2\times 10^2+10\times 10^1 +3\times 10^0$ болно. Энэ хэлбэрээр илэрхийлэх аргыг 10 суурьтай гэж хэлдэг. Тоог дурын өөр суурь тоогоор мөн илэрхийлж болох буюу 2-тын тооллын суурь авч үзэе. 2-тйн тооллын суурь гэдэг нь дурын тоог 0 болон 1 ээр илэрхийлэх юм. 9213-ыг 2тын тооллын сууриар илэрхийлбэл $10001111111101$ болох буюу өөрөөр хэлбэл $$ 9213=1\times 2^{13}+0\times 2^{12}+0\times 2^{11}+0\times 2^{10}+ 
1\times 2^{9}+1\times 2^{8}+1\times 2^{7}
\\+1\times 2^{6}+1\times 2^{5}+1\times 2^{4}+1\times 2^{3}+1\times 2^{2}+
0\times 2^{1}+1\times 2^{0}.$$ 
2-тын тооллын системийг ашиглан дурын тооноос гадна үг, өгүүлбэр, зураг, дүрс ба видео гэх мэтчилэн компьютерийн дэлгэц дээр харагдаж буй бүх зүйлсийг илэрхийлж болно. Цаашилбал бидний өдөр тутамд хэрэглэдэг компьютер нь бүх мэдээллийг 2-тын тооллын системээр илэрхийлдэг. Бидний цаашид үзэх квант компьютер ердийн компьютертэй ижил зарчимтай. Зарим нэг ялгаа нь гэвэл квант компьютерийн бит (bit)-ийг квант бит (quantum bit) буюу күбит (qubit) гэж нэрлэдэг.
$\textbf{2-тын тооллын системийн талаарх дэлгэрэнгүй мэдээллийг:}$ https://web.stanford.edu/class/cs101/bits-bytes.html

# Таны анхны квант компьютерийн хэлхээ

```python
#Хэрэглэх package
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram 
```

Бидний авч үзэх квант хэлхээ бол 8-н күбитийн хэлхээ юм. Квант тооцоолол гүйцэтгэх энгийн 3н алхам байдаг. 1-рт оролтын мэдээллээ энкодлох буюу битээр илэрхийлэх, 2-рт нь ямар нэгэн тооцооллыг гүйцэтгэх, 3-рт нь үр дүн битээс сүүлийн мэдээллээ гаргаж (extraction of information/outputs) авах.  

```python
n=8
n_q=8
n_b=8
qc_output=QuantumCircuit(n_q,n_b)
```

Квант хэлхээнээс гаралтын мэдээллийг авах үйлдлийг "хэмжүүр (measure)" гэж хэлдэг. Хэмжүүр гэдэг нь күбитийг бит үрүү буулгах үйлдэл юм. Өөр үгээр илэрхийлбэл квант компьютер (qc) дээр гүйцэтгэсэн хэмжилт буюу хэмжүүр (measure) маань квант бит (j)-г харгалзах классик бит (j)-д буулгах (qc.measure(j,j)).

```python
#хэмжүүр: (күбитийг-бит үрүү) буулгах үйлдэл. measure: (qubit to bit) mapping operation. 
for j in range(n):
    qc_output.measure(j,j)
```

Квант хэлхээгээ зурж үзэцгээе.

```python
qc_output.draw(output='mpl')
```

Күбитийн анхны утга нь 0 бол мөн гаралтын утга нь үргэлж 0 байхаар тохируулсан физик систем авч үзэж байгаа болно. Ийм системийн хувьд гаралтын үр дүнг нь 0 байна гэдгийг хистограмм график байгуулан харж болно. 

```python
counts=execute(qc_output, Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

Өмнө дурдсанчлан квант компьютер нь квант физикийн хачирхалтай физик үзэгдэлд суурилдаг төдийгүй, хэмжилтийн үр дүн (measure) тодорхойгүй байдаг (indetermenistic or not deterministic) ба үргэлж хэмжилтийн үр дүн санамсаргүй байдаг. Гэвч дээрх хэмжилтийн үр дүн маш тодорхой (deterministic) буюу хэмжилтийг хэдэн үдаа давтан хийснээс үл харгалзан нэг л утга буюу 00000000 байна. Учир нь дээр дурдсанчлан бид бүхэн күбитээ гадны ямар нэгэн нөөлөөнөөс тусгаарласан байгаа буюу квант шинж чанар нь гарч ирэх ямар нэгэн квант үйлдэл (quantum operation) хийгээгүй болно.    

# Оролтын мэдээллийг энкодлох (encoding) буюу 2-тын тооллын системд илэрхийлэх

Энэ хэсэгт NOT gate-ийн үндсэн үйлдийн талаар авч үзэцгээе. NOT gate нь 0-ийг 1 харин 1-ийг 0 болгодог үйлдэл бөгөөд квант компьютерт энэ үйлдэлтэй ижилхэн үүрэгтэй gate-ийг Х gate гэж нэрлэдэг. X gate-ийг дээр дурдсан жишээний хамгийн сүүлийн күбит дээр гүйцэтгэе.

```python
qc_encode=QuantumCircuit(n)
qc_encode.x(7)
qc_encode.draw(output='mpl')
```

```python
#хэмжилтийн үр дүнгээ харцгаая
qc=qc_encode+qc_output
qc.draw(output='mpl',justify='none')
```

```python
counts = execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

Эндээс харвал сүүлийн күбит маань хамгийн эхэнд бичигдсэн байна. Шалтгаан нь гэвэл qiskit програм нь сүүлээс нь эхэлж дугаарладаг юм. $10000000$ күбит нв 128 гэдэг тоонд харгалзаж байна буюу бид х gate ашиглаж квант компьютер дээрээ 128 тоог бичлээ.

# Qiskit ашиглан нэмэх үйлдэл гүйцэтгэе

```python
qc_cnot = QuantumCircuit(2)
qc_cnot.cx(0,1)
qc_cnot.draw(output='mpl')
```

```python
qc = QuantumCircuit(2,2)
qc.x(0)
qc.cx(0,1)
qc.measure(0,0)
qc.measure(1,1)
qc.draw(output='mpl')
```

```python
qc_ha = QuantumCircuit(4,2)
# encode inputs in qubits 0 and 1
qc_ha.x(0) # For a=0, remove this line. For a=1, leave it.
qc_ha.x(1) # For b=0, remove this line. For b=1, leave it.
qc_ha.barrier()
# use cnots to write the XOR of the inputs on qubit 2
qc_ha.cx(0,2)
qc_ha.cx(1,2)
qc_ha.barrier()
# extract outputs
qc_ha.measure(2,0) # extract XOR value
qc_ha.measure(3,0)

qc_ha.draw(output='mpl')
```

```python
qc_ha = QuantumCircuit(4,2)
# encode inputs in qubits 0 and 1
qc_ha.x(0) # For a=0, remove the this line. For a=1, leave it.
qc_ha.x(1) # For b=0, remove the this line. For b=1, leave it.
qc_ha.barrier()
# use cnots to write the XOR of the inputs on qubit 2
qc_ha.cx(0,2)
qc_ha.cx(1,2)
# use ccx to write the AND of the inputs on qubit 3
qc_ha.ccx(0,1,3)
qc_ha.barrier()
# extract outputs
qc_ha.measure(2,0) # extract XOR value
qc_ha.measure(3,1) # extract AND value

qc_ha.draw(output='mpl')
```

```python
counts = execute(qc_ha,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```
